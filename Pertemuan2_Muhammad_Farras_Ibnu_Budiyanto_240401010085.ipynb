```python
# =======================================================
# MEMUAT DATASET TITANIC DAN MENJAWAB Q2
# =======================================================
import numpy as np
import pandas as pd

# 1. pd.read_csv() fungsinya untuk membaca file tabel/csv dari link internet
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
display(df)

# .isnull().sum() akan menghitung berapa total baris yang kosong di tiap kolom
print("\n=== 1. DATA KOSONG ===")
data_kosong = df.isnull().sum()
print(data_kosong)

print("\n=== 2. PERSENTASE DATA KOSONG (BIAR LEBIH SPESIFIK) ===")
# Menghitung persentase data bolong: (jumlah kosong / total seluruh baris) * 100
persen_kosong = (df.isnull().sum() / len(df)) * 100
print(persen_kosong.round(2).astype(str) + ' %')

# df.isnull().sum() digunkan untuk mencari jumlah nilai kosong
# menggunakan operasi jumlah data bolong/jumlah baris * 100 untuk mencari rata-rata
# .round(2) digunakan untuk nilai desimal cukup membawa 2 nilai (3.14), dan astype(str) agar data dalam bentuk angka bisa digabung dengan data string
```
